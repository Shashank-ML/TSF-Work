## Visualising and analysing data

```python
corelation = df.corr()
```

```python
# visulaising correlations using heat map
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot =True)
```

```python
# finding relations in pairs
sns.pairplot(df)
```

```python
# plotting relation between profit and sales on the basis of discount
sns.relplot(x='Profit', y = 'Sales', hue = "Discount",data = df)
```

```python
# plotting the profits and sales of every category
sns.relplot(x='Profit', y = 'Sales', hue = "Category", data = df )
```

```python
# analysing profits
sns.distplot(df['Profit'])
```

```python
#Sub category wise
fig = plt.figure(figsize= (10,7))
df['Sub-Category'].value_counts().plot.pie(autopct = '%.2f',shadow = False)
```

```python
#profits
profit = df['Profit']
profit
```

```python
# analysing losses
loss = df[df['Profit']< 0]
loss
```

```python
#identifying maximum profit
max_profit =  df.loc[df['Profit'].idxmax()]
max_profit
```

```python
#discount wise profit
discount_wise = df.groupby('Discount').sum()['Profit']
print("Discount wise Profit in Ascending Order: \n", discount_wise)
```

```python
#plotting the discount and profit relation
sns.catplot(x="Discount", y="Profit", kind="bar", data=df)
```

```python
# visualising region wise profit
sns.catplot(x="Region", y="Profit", kind="bar", data=df)
```

```python
State_Profit = df.groupby('State').sum()['Profit']
print("State wise profit:\n",State_Profit)
```

```python
# state wise profit
State_Profit.sort_values(ascending = False)
#plot
plt.figure(figsize =(18,7))
sns.barplot(x = State_Profit.index, y = State_Profit.values)
plt.xticks(rotation =90)
plt.title("State Wise Profit")
```

```python
# city wise profit
City_Profit = df.groupby('City').sum()['Profit']
print("City wise Profit in Ascending Order: \n", City_Profit)
```

```python
# plotting top 5 cities and worst 5 cities in terms of profit 
city_sorted1 = City_Profit.sort_values(ascending = False).head()
#plot
fig,axes = plt.subplots(1,2, figsize = (15,5))
city_sorted1.head().plot(kind ='bar', ax = axes[0])
axes[0].tick_params(labelrotation = 90)
axes[0].set_title("Maximum City Profit (Top 5)")
city_sorted2 = City_Profit.sort_values(ascending = True).head()
city_sorted2.head().plot(kind='bar',ax = axes[1])
plt.xticks(rotation = 90)
plt.title("Minimum City Profit (Bottom 5)")
```

```python
# relation between sales and segment
sns.catplot(x="Segment", y="Sales", kind="bar", data=df)
```

```python
sns.catplot(x="Segment", y="Profit", kind="bar", data=df)
```

```python
# analysing relation between shipmode and profit/sales
sns.catplot(x="Ship Mode", y="Sales", kind="bar", data=df)
```

```python
sns.catplot(x="Ship Mode", y="Profit", kind="bar", data=df)
```

```python
# Category wise profit and sales
sns.catplot(x="Category", y="Profit", kind="bar", data=df)
```

```python
sns.catplot(x="Category", y="Sales", kind="bar", data=df)
```

```python
# Sub Cateogry wise profit and sales
Sub_Category_Profit = df.groupby('Sub-Category').sum()['Profit']
print("Sub-Category wise Profit:\n",Sub_Category_Profit)
```

```python
Sub_Category_Profit.sort_values(ascending=False)
#plot
plt.figure(figsize =(18,7))
sns.barplot(x = Sub_Category_Profit.index, y = Sub_Category_Profit.values)
plt.xticks(rotation =90)
plt.title("Sub-category wise Profit")
```
